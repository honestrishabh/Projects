```python
import pandas as pd
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_squared_error as mse
from sklearn.metrics import mean_absolute_error as mae
from sklearn.metrics import root_mean_squared_error as rmse
import matplotlib.pyplot as plt

from tqdm import tqdm
#ORDER

pdq_order = (1, 1, 3)  # Non-seasonal order (p, d, q)
PDQs_order = (1, 1, 2, 12)  # Seasonal order (P, D, Q, s)

def forecast_pfc(pfc_data):

  
  model = SARIMAX(pfc_data, order=pdq_order, seasonal_order=PDQs_order)
  model_fit = model.fit()


  pred_uc = model_fit.get_forecast(steps=24)
  pred_ci = pred_uc.conf_int()

  # Plot
  plt.figure(figsize=(10, 6))  
  pfc_data.plot(label='Observed Sales')
  pred_uc.predicted_mean.plot(label='Forecasted Sales')
  plt.fill_between(pred_ci.index, pred_ci.iloc[:, 0], pred_ci.iloc[:, 1], color='k', alpha=0.25, label='Confidence Interval')
  plt.xlabel('Date')
  plt.ylabel('Sales(in crores)')
  plt.title(f"Forecast for PFC {pfc_data.name}")  
  plt.legend()
  plt.grid(True)
  plt.tight_layout()  
  plt.show()  

  return pred_uc.predicted_mean  # Return forecasts 


forecasts = {}
forecast_df=pd.DataFrame()
with tqdm(total=len(df1.columns)) as pbar:  # Adjust total based on expected iterations
    for pfc_id in df1.columns:
        pfc_data = df1[pfc_id]
        forecast = forecast_pfc(pfc_data)
        forecasts[pfc_id] = forecast
        forecast_df[pfc_id] = forecast
        pbar.update(1)  # Update progress bar for each iteration

        # Model Evaluation (accuracy)
        actual_values = pfc_data.iloc[-len(forecast):]
        acc_value = rmse(actual_values, forecast)
        print(f"PFC ID: {pfc_id}, MAE:{acc_value}")

print("Forecasts for all PFCs:")
for pfc_id, forecast in forecasts.items():
    print(f"PFC ID: {pfc_id}, Forecast: {forecast.tolist()}")
  
```

```python
import pandas as pd
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_squared_error as mse
from sklearn.metrics import mean_absolute_error as mae
import matplotlib.pyplot as plt

# Define the non-seasonal order and seasonal order for the SARIMA model
pdq_order = (0, 1, 3)  # Non-seasonal order (p, d, q)
PDQs_order = (1, 1, 2,12)  # Seasonal order (P, D, Q, s)

def forecast_pfc(pfc_data):
  # Train the SARIMA model on the PFC data
  model = SARIMAX(pfc_data, order=pdq_order, seasonal_order=PDQs_order)
  model_fit = model.fit()

  # Generate predictions for future periods (replace 12 with your desired forecast horizon)
  preds = model_fit.predict(steps=12)
  return preds


forecasts = {}
for pfc_id in df1.columns:  # Iterate through column names (former PFC IDs)
  pfc_data = df1[pfc_id]  # Extract the sales data for this PFC (now a Series)
  forecast = forecast_pfc(pfc_data)
  forecasts[pfc_id] = forecast

  # Evaluate forecast accuracy (optional)
  actual_values = pfc_data.iloc[-len(forecast):]  # Get actual values for the forecast period
  acc_value =mae(actual_values, forecast)
  

  # Plot forecast vs actual values
  plt.figure(figsize=(10, 6))  # Set plot size
  plt.plot(pfc_data.index, pfc_data.values, label='Actual Sales')
  plt.plot(pfc_data.index[-len(forecast):], forecast.values, label='Forecast')
  plt.xlabel('Date')
  plt.ylabel('Sales')
  plt.title(f"Forecast for PFC {pfc_id}")
  plt.legend()
  plt.grid(True)
  plt.show()  # Display the plot

# Use the forecasts dictionary for further analysis

print("Forecasts for all PFCs:")
for pfc_id, forecast in forecasts.items():
  print(f"PFC ID: {pfc_id},MAE: {acc_value:.2f} ,Forecast: {forecast.tolist()}")
  # print(f"PFC ID: {pfc_id}, )  # Print Mean Squared Error
```
